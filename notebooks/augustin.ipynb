```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data_path = "../data/"
```

# Target variable

```python
df = pd.read_csv(data_path + "KPI_to_model.csv")
df.head()
```

```python
# We focus on units sold not on revenue
df_focus = df.drop(columns=[
    "UK L'Oreal Paris Haircare Total Offline Sellout Value (in pound)",
    "UK L'Oreal Paris Haircare Total Online Sellout Value (in pound)"])
df_focus["Starting Week"] = pd.to_datetime(df_focus["Starting Week"])
```

```python
# Let us plot the online and offline sales
sns.lineplot(data=df_focus, x="Starting Week", y="UK L'Oreal Paris Haircare Total Online Sellout Units", label="Online")
sns.lineplot(data=df_focus, x="Starting Week", y="UK L'Oreal Paris Haircare Total Offline Sellout Units", label="Offline")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.lineplot(data=df_focus, x="Starting Week", y="UK L'Oreal Paris Haircare Total Online Sellout Units", label="Online")
plt.xticks(rotation=45, ha='right')
plt.title('Online Sales')
plt.show()
```

```python
sns.lineplot(data=df_focus, x="Starting Week", y="UK L'Oreal Paris Haircare Total Offline Sellout Units", label="Offline", color='orange')
plt.xticks(rotation=45, ha='right')
plt.title("Offline Sales")
plt.show()
```

```python
# Let us plot a normalized version of the data
df_focus["Online"] = df_focus["UK L'Oreal Paris Haircare Total Online Sellout Units"] / df_focus["UK L'Oreal Paris Haircare Total Online Sellout Units"].max()
df_focus["Offline"] = df_focus["UK L'Oreal Paris Haircare Total Offline Sellout Units"] / df_focus["UK L'Oreal Paris Haircare Total Offline Sellout Units"].max()

sns.lineplot(data=df_focus, x="Starting Week", y="Online", label="Online")
sns.lineplot(data=df_focus, x="Starting Week", y="Offline", label="Offline")
plt.xticks(rotation=45, ha='right')
plt.show()
```
